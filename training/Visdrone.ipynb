```python
import os
import subprocess

local_path = os.path.expanduser("../gdrive/YOLOv10")
if not os.path.exists(local_path):
    print(f"Creating local directory: {local_path}")
    os.makedirs(local_path, exist_ok=True)
else:
    print(f"Using local directory: {local_path}")

print(
    "To update from Google Drive, run: rclone sync gdrive:YOLOv10 ~/gdrive/YOLOv10 --transfers=32 --checkers=32 --filter-from filter-rules.txt -P"
)
print("Syncing YOLOv10 folder from Google Drive (this may take a while)...")
result = subprocess.run(
    [
        "rclone",
        "sync",
        "gdrive:YOLOv10",
        local_path,
        "--transfers=32",
        "--checkers=32",
        "--filter-from",
        "filter-rules.txt",
        "-P",
    ],
    capture_output=False,
)

if result.returncode == 0:
    print("✓ Sync completed successfully")
else:
    print("✗ Sync failed - please check rclone configuration")
```

```python
from ultralytics import YOLO

# Set paths for local training
pre_trained_path = os.path.join(local_path, "yolo12l.pt")
# lastest_pre_trained_path = os.path.join(local_path, "weights/pre-trained/160_epochs_visdrone_v12/weights/last.pt")
data_path = os.path.join(local_path, "data_train/VisDrone/VisDrone_no_download.yaml")
name = "160_epochs_visdrone_v12"
project = os.path.join(local_path, "weights/pre-trained/")

print(f"Data path: {data_path}")
print(f"Project path: {project}")

# Load a pretrained model
model = YOLO(pre_trained_path)
# model = YOLO(lastest_pre_trained_path)

# Train the model
results = model.train(
    data=data_path,
    project=project,
    name=name,
    epochs=160,
    imgsz=640,
    batch=8,
    device="mps",
)
# results = model.train(data=data_path, project=project, name=name, epochs=160, imgsz=640, batch=12, resume=True)
```
